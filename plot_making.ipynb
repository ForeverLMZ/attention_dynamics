```python
import numpy as np
import matplotlib.pyplot as plt

# Load the files with allow_pickle=True
train_losses = np.load('./train_losses.npy', allow_pickle=True)
train_accuracies = np.load('./train_accuracies.npy', allow_pickle=True)
test_losses = np.load('./test_losses.npy', allow_pickle=True)
test_accuracies = np.load('./test_accuracies.npy', allow_pickle=True)
```

```python
import numpy as np

file_path = './train_accuracies.npy'

# Check the file using numpy.lib.format.open_memmap
_ = np.lib.format.open_memmap(file_path)
```

```python
# Plot the training and testing loss
plt.figure(figsize=(10, 5))
x = range(len(train_losses))
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.plot(train_losses, label='Training Loss')
plt.plot(test_losses, label='Testing Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Testing Loss')
plt.legend()
plt.show()
```

```python
# Plot the training and testing accuracy
plt.figure(figsize=(10, 5))
x = range(len(train_losses))
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.plot(train_accuracies, label='Training Accuracy')
plt.plot(test_accuracies, label='Testing Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Model Performance')
plt.legend()
plt.show()
```
